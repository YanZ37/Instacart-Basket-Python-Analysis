### Step 2. Consistency checks on orders.csv dataset

```python
# Import orders_wrangled.csv dataset
df_ords = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_wrangled.csv'), index_col = False)
```

```python
# Change the column name from 'order_dow' to 'order_day_of_week'
df_ords.rename(columns = {'order_dow' : 'order_day_of_week'}, inplace = True)
```

#### 2. Check the max and min values for orders_wrangled.csv dataset

```python
# Generates summary statistics of the dataframe
df_ords.describe()
```

#### 3. Check for mixed-type data in df_ords dataframe.

```python
for col in df_ords.columns.tolist():
  weird = (df_ords[[col]].map(type) != df_ords[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_ords[weird]) > 0:
    print (col)
```

#### **Summary**: No mixed data type was found in df_ords dataframe.

#### 5. Run a check for missing values in df_ords dataframe

```python
# Identify missing values for df_ords.csv
df_ords.isnull().sum()
```

#### **Summary**: The number of missing values for 'days_since_prior_order' is 206,209.

```python
# View the columns that contain the missing values
df_ords_nan = df_ords[df_ords['days_since_prior_order'].isnull() == True]
```

```python
df_ords_nan
```

```python
# Examine whether the null values in 'days_since_prior_order' are because they are the 1st orders
df_ords_first = df_ords_nan[df_ords_nan['order_number'] == 1]
```

```python
df_ords_first
```

#### **Summary:** The row numbers of "'order number' = 1" matches the row number "'days_since_prior_order' is null". So we can conclude that the reasons for the null values in 'days_since_prior_order'column are because they are the first orders.

#### 6. Address the missing values: Flag the missing values in 'days_since_prior_order' in a new column.


```python
df_ords['first_order'] = np.where(df_ords['order_number'] == 1, True, False)
```
