### **Consistency checks**

#### 2. Check for mixed-type data in df_cus dataframe.

```python
for col in df_cus.columns.tolist():
  weird = (df_cus[[col]].map(type) != df_cus[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cus[weird]) > 0:
    print (col)
```

#### **Observation:** 
'first_name' variable contains mixed-datatype data.

#### Address the mixed datatype

```python
# Change all datatypes in first_name to string
df_cus['first_name'] = df_cus['first_name'].astype('str')
```

#### 5. Merge the datasets

```python
# Inspect the column names
ords_prods_merged_new.columns
```

```python
# Delete the previous merging label '_merge'
df1 = ords_prods_merged_new.drop('_merge', axis = 1)
```

```python
# Inspect the descriptive statistics of df1
df1.describe()
```

```python
# Merge df_cus and df1 using user_id as a key, add a merge flag
merged = pd.merge(df_cus, df1, on = 'user_id', how = 'inner', indicator = True)
```

```python
# Inspect the merging flag
merged['_merge'].describe()
```

```python
# Export to a pickle
merged.to_pickle(os.path.join(path, '02 Data','Prepared Data','merge.pkl'))
```
