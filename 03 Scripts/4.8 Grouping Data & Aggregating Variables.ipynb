### Step 1. Import the libraries and datasets

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Create path - Shortcut for Importing Data Files
path = r'C:\Users\sanch\Documents\Instacart Basket Analysis'
```

```python
# Import ords_prods_merge.pkl dataset
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_updated.pkl'))
```

```python
ords_prods_merge.head()
```

```python
#### Step 2. Find the aggregated mean of the “order_number” column grouped by “user_id”
```

```python
# Return the average order numbers for each department
ords_prods_merge.groupby('user_id').agg({'order_number': ['mean']}).describe()
```

```python
# Create a column 'avg_order' by generating the mean order numbers for each customer
ords_prods_merge['avg_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.mean)
```

```python
# Reset Panda options regarding the maximum number of rows to display
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head()
```

```python
# Define the range for 'Loyal customer', 'Regular customer' & 'New customer'
ords_prods_merge.loc[ords_prods_merge['avg_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['avg_order'] <= 40) & (ords_prods_merge['avg_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['avg_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# View the new columns
ords_prods_merge[['user_id', 'order_number', 'loyalty_flag', ]].head()
```

```python
ords_prods_merge.head()
```

### Step 3. Find the aggregated average price for each user

```python
# Return the average prices for each user group
ords_prods_merge.groupby('user_id').agg({'prices': ['mean']}).head()
```

```python
# Create a column 'avg_prices' by generating the average prices for each user
ords_prods_merge['avg_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Define the range for 'Low spender' & 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.head()
```

```python
# View three columns 'user_id', 'avg_prices' & 'spending_flag'
ords_prods_merge[['user_id', 'avg_prices', 'spending_flag']].head()
```

### Step 4. Find the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. 

```python
# Return the median 'days_since_prior_order' for each user
ords_prods_merge.groupby('user_id').agg({'days_since_prior_order': ['median']}).head()
```

```python
# Create a column 'median_days_prior' by generating the median 'days_since_prior_order' for each user
ords_prods_merge['median_days_prior'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Define 'Non-frequent customer', 'Regular customer' & 'Frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_days_prior'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_prior'] <= 20) & (ords_prods_merge['median_days_prior'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_prior'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# View three columns 'user_id', 'median_days_prior', 'frequency_flag'
ords_prods_merge[['user_id', 'median_days_prior', 'frequency_flag']].head()
```
